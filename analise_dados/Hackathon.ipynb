```python
# lê o arquivo CSV
#df = pd.read_csv('../dados/metadados2020.csv.gz',encoding='latin-1')

import pandas as pd
import glob

path = r'../dados/' # use your path
all_files = glob.glob(path + "/*.csv.gz")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0,encoding='latin-1')
    li.append(df)

#df = pd.concat((pd.read_csv(f) for f in all_files))
    
frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
frame.columns = ['Número do processo','Código da classe processual','Descrição da classe processual','Código do Assunto','Descrição do Assunto','Nome Parte','Polo Ativo ou Passivo','Tipo de Parte','Data da distribuição','Órgão Julgador','Data Movimentação Processual','Código do Evento','Descrição do Evento','Descrição do Complemento','Documento','Conteúdo do Documento','Magistrado responsável pelo ato','Valor da Causa','Processo Originário','justica_gratuita','Processo Prioritário','Tutela/Luminar','Apreciação Tutela/Liminar','Data Trânsito em Julgado','Processo Referência','Revisado','Incidental','Data de Autuação']
# Ajusta Dtype das colunas de data
frame['Data Trânsito em Julgado'] = pd.to_datetime(frame['Data Trânsito em Julgado'])
frame['Data Movimentação Processual'] = pd.to_datetime(frame['Data Movimentação Processual'])
frame['Data da distribuição'] = pd.to_datetime(frame['Data da distribuição'])
frame['Data de Autuação'] = pd.to_datetime(frame['Data de Autuação'])
frame['Data Movimentação Processual'] = pd.to_datetime(frame['Data de Autuação'])
```

```python
frame
```

```python
# Ajuste dos valores booleanos para True e False
frame['justica_gratuita'].replace(['S','N'],[1,0],inplace=True)
frame['Processo Prioritário'].replace(['S','N'],[1,0],inplace=True)
frame['Tutela/Luminar'].replace(['S','N'],[1,0],inplace=True)
frame['Revisado'].replace(['S','N'],[1,0],inplace=True)
frame['Incidental'].replace(['S','N'],[1,0],inplace=True)
frame['Apreciação Tutela/Liminar'].replace(['S','N'],[1,0],inplace=True)

# Ajusta Dtype das colunas de data
frame['justica_gratuita'] = frame['justica_gratuita'].astype('bool')
frame['Processo Prioritário'] = frame['Processo Prioritário'].astype('bool')
frame['Tutela/Luminar'] = frame['Tutela/Luminar'].astype('bool')
frame['Revisado'] = frame['Revisado'].astype('bool')
frame['Incidental'] = frame['Incidental'].astype('bool')
frame['Apreciação Tutela/Liminar'] = frame['Apreciação Tutela/Liminar'].astype('bool')

```
